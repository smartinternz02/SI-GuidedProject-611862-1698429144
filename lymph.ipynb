```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
#feature scaling
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled
```

```python
#Splitting Data into Train and Test.
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
x_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
params={
    'max_depth':[9,10,11],
    'min_samples_leaf':[2,3],
    'n_estimators':[90,95,100,110],
    'max_features':[2,3,4,5]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search=GridSearchCV(estimator=rf,
                        param_grid=params,
                        cv=2,
                        verbose=1,
                        scoring="accuracy")
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_score_
```

```python
rf_best=grid_search.best_estimator_
rf_best
```

```python
rf_classify=RandomForestClassifier(random_state=42,
                                   n_jobs=-1,
                                   max_depth=9,
                                   min_samples_split=2,
                                   max_features='sqrt',
                                   n_estimators=90,
                                   bootstrap=True)
```

```python
rf_classify.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
prediction=rf_classify.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,prediction)
```

```python
accuracy_score(y_test,prediction)
```

```python
print(classification_report(y_test,prediction))
```

```python
import pickle

pickle.dump(rf_classify,open('saved_model.pkl','wb'))
pickle.dump(ms,open('ms_saved.pkl','wb'))
```
